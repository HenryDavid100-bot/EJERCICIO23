```python
crear una calculadora
# Funciones para cada operación
def sumar(x, y):
    return x + y

def restar(x, y):
    return x - y

def multiplicar(x, y):
    return x * y

def dividir(x, y):
    if y != 0:
        return x / y
    else:
        return "Error: No se puede dividir por 0."

# Mostrar las opciones
def mostrar_opciones():
    print("Operaciones disponibles:")
    print("1. Sumar")
    print("2. Restar")
    print("3. Multiplicar")
    print("4. Dividir")

# Función principal de la calculadora
def calculadora():
    while True:
        mostrar_opciones()

        # Solicitar la opción del usuario
        try:
            seleccion = int(input("Selecciona una operación (1/2/3/4): "))
        except ValueError:
            print("Por favor, ingresa un número válido.")
            continue

        # Verificar si la opción es válida
        if seleccion not in (1, 2, 3, 4):
            print("Opción no válida, intenta de nuevo.")
            continue

        # Solicitar los números
        try:
            num1 = float(input("Ingresa el primer número: "))
            num2 = float(input("Ingresa el segundo número: "))
        except ValueError:
            print("Por favor, ingresa un número válido.")
            continue

        # Realizar la operación según la opción seleccionada
        if seleccion == 1:
            print(f"{num1} + {num2} = {sumar(num1, num2)}")
        elif seleccion == 2:
            print(f"{num1} - {num2} = {restar(num1, num2)}")
        elif seleccion == 3:
            print(f"{num1} * {num2} = {multiplicar(num1, num2)}")
        elif seleccion == 4:
            print(f"{num1} / {num2} = {dividir(num1, num2)}")

        # Preguntar si el usuario quiere continuar
        continuar = input("¿Quieres hacer otra operación? (sí/no): ").strip().lower()
        if continuar != "sí":
            print("Gracias por usar la calculadora. ¡Hasta luego!")
            break

# Llamar a la función principal
calculadora()


Calculadora cientifica
import math

# Funciones para cada operación
def suma(x, y):
    return x + y

def resta(x, y):
    return x - y

def multiplicar(x, y):
    return x * y

def dividir(x, y):
    if y != 0:
        return x / y
    else:
        return "Error: No se puede dividir por 0."

def potencia(x, y):
    return x ** y

def raiz(x):
    if x >= 0:
        return math.sqrt(x)
    else:
        return "Error: La raíz cuadrada no está definida para números negativos."

def logaritmo(x, base=math.e):
    if x > 0:
        return math.log(x, base)
    else:
        return "Error: El logaritmo no está definido para números menores o iguales a cero."

def seno(x):
    return math.sin(math.radians(x))

def coseno(x):
    return math.cos(math.radians(x))

def tangente(x):
    return math.tan(math.radians(x))

# Mostrar las opciones
def mostrar_opciones():
    print("Operaciones disponibles:")
    print("1. Sumar")
    print("2. Restar")
    print("3. Multiplicar")
    print("4. Dividir")
    print("5. Potencia")
    print("6. Raíz cuadrada")
    print("7. Logaritmo")
    print("8. Seno")
    print("9. Coseno")
    print("10. Tangente")
    print("11. Salir")

# Función principal de la calculadora científica
def calculadora_cientifica():
    while True:
        mostrar_opciones()

        # Solicitar la opción del usuario
        try:
            seleccion = int(input("Selecciona una operación (1-11): "))
        except ValueError:
            print("Por favor, ingresa un número válido.")
            continue

        # Salir si elige la opción 11
        if seleccion == 11:
            print("Gracias por usar la calculadora científica. ¡Hasta luego!")
            break

        # Realizar la operación según la opción seleccionada
        if seleccion == 1:
            num1 = float(input("Ingresa el primer número: "))
            num2 = float(input("Ingresa el segundo número: "))
            print(f"{num1} + {num2} = {suma(num1, num2)}")
        elif seleccion == 2:
            num1 = float(input("Ingresa el primer número: "))
            num2 = float(input("Ingresa el segundo número: "))
            print(f"{num1} - {num2} = {resta(num1, num2)}")
        elif seleccion == 3:
            num1 = float(input("Ingresa el primer número: "))
            num2 = float(input("Ingresa el segundo número: "))
            print(f"{num1} * {num2} = {multiplicar(num1, num2)}")
        elif seleccion == 4:
            num1 = float(input("Ingresa el primer número: "))
            num2 = float(input("Ingresa el segundo número: "))
            print(f"{num1} / {num2} = {dividir(num1, num2)}")
        elif seleccion == 5:
            num1 = float(input("Ingresa la base: "))
            num2 = float(input("Ingresa el exponente: "))
            print(f"{num1} ^ {num2} = {potencia(num1, num2)}")
        elif seleccion == 6:
            num = float(input("Ingresa el número para calcular la raíz cuadrada: "))
            print(f"La raíz cuadrada de {num} es {raiz(num)}")
        elif seleccion == 7:
            num = float(input("Ingresa el número para calcular el logaritmo: "))
            base = input("Ingresa la base (por defecto es 'e'): ")
            base = float(base) if base else math.e
            print(f"El logaritmo de {num} en base {base} es {logaritmo(num, base)}")
        elif seleccion == 8:
            angulo = float(input("Ingresa el ángulo en grados para calcular el seno: "))
            print(f"Seno de {angulo}° = {seno(angulo)}")
        elif seleccion == 9:
            angulo = float(input("Ingresa el ángulo en grados para calcular el coseno: "))
            print(f"Coseno de {angulo}° = {coseno(angulo)}")
        elif seleccion == 10:
            angulo = float(input("Ingresa el ángulo en grados para calcular la tangente: "))
            print(f"Tangente de {angulo}° = {tangente(angulo)}")
        else:
            print("Opción no válida. Intenta de nuevo.")

# Llamar a la función principal
calculadora_cientifica()
```
